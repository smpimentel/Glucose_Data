```python
import pandas as pd
import io
from datetime import date
today = date.today()
date = today.strftime("%m-%d-%Y") 
data = f'uncleanded_csv/StephenPimentel_glucose_{date}.csv'
df = pd.read_csv(data)
df = pd.read_csv(io.StringIO(u""+df.to_csv(header=None)))
df = df[['Device Timestamp', 'Historic Glucose mg/dL']]
df = df.dropna()
df.rename(columns={"Device Timestamp":"Time", "Historic Glucose mg/dL":"Glucose"}, inplace=True)
df.to_json(r'cleaned_csv/cleaned_data.json')
```
